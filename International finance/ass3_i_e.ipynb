Home assignment 3: Comparative Advantage

Belgium has L = 100 and Holland has L_h = 200. Both countries produce two goods: beer and tulips. 
In Belgium: MPLB=1 and MPLT=1/2
The production technology in Holland is MPL_B_H=1/2 and MPL_T_H=1

Also assume that consumers in both countries have Leontief preferences, consuming B and T in the fixed proportion of one-to-one.

(a) Which country has a comparative advantage in which good?

```python
import matplotlib.pyplot as plt
import numpy as np


Belgium_L = 100
Holland_L = 200
MPLB_B = 1
MPLT_B = 1/2
MPL_B_H = 1/2
MPL_T_H = 1

# calculate the opportunity cost of producing one unit of beer in Belgium
OCB = MPLT_B/MPLB_B

# calculate the opportunity cost of producing one unit of beer in Holland
OC_B_H = MPL_T_H/MPL_B_H

# calculate the opportunity cost of producing one unit of tulips in Belgium
OCT = MPLB_B/MPLT_B

# calculate the opportunity cost of producing one unit of tulips in Holland

OC_T_H = MPL_B_H/MPL_T_H

print(f'Belgium beer: {OCB}, Holland beer: {OC_B_H},Belgium tulips: {OCT}, Holland tulips: {OC_T_H}', sep='\n')

# Now calculate the opportunity costs of beer in belgium
# Opt_cost_tulip_belg = OCB/OCT
# Opt_cost_beer_belg = OCT/OCB


# Opt_cost_beer_holland = OC_T_H/OC_B_H
# Opt_cost_tulips_holland = OC_B_H/OC_T_H

# print(f'Holland beer: {Opt_cost_beer_holland}, Holland tulips: {Opt_cost_tulips_holland}', sep='\n')
# print(f'Belgium beer: {Opt_cost_beer_belg}, Belgium tulips: {Opt_cost_tulip_belg}', sep='\n')
```

Belgium has a lower opportunity cost for producing beer compared to Holland (0.5<2). Therefore, Belgium has a comparative advantage in producing beer.

Holland has a lower opportunity cost for producing tulips compared to Belgium (0.5<2). Therefore, Holland has a comparative advantage in producing tulips.


Draw the production possibilities frontier for each country. 

(b) What is the relative price of beer in terms of tulips in each country?

```python
# Draw the production possibilities frontier for each country. Called  the PPF
# First we make a function for the slope of the PPF which is the derivative of beer with respect to tulips

def PPF_B(x):
    return Belgium_L - x

def PPF_H(x):
    return Holland_L - x

# We can now plot the PPF for each country
x = np.linspace(0, 100, 100)
plt.plot(x, PPF_B(x))
plt.xlabel('Tulips')
plt.ylabel('Beer')
plt.title('Belgium PPF');
```

```python
# Holland
x = np.linspace(0, 200, 200)
fig, ax = plt.subplots()
ax.plot(x, PPF_H(x), label='Holland')
# title
plt.title('Holland');
```

```python
import sympy as sp
# How many units of B and T will be produced and consumed in each country with leontif preferences?
B, T = sp.symbols('B T')  # Beer and Tulips

# Belgium PPT
equ = sp.Eq(B/MPLB_B+B/MPLT_B, Belgium_L)

solve_beer_belgium = sp.solve(equ, B)
# They are the same, so we can use the same variable to store the result
# So if beer is 33.33, then tulips will be 33.33 as well
solve_tulips_belgium = solve_beer_belgium

# Holland PPT
equ_h = sp.Eq(T/MPL_B_H + T/MPL_T_H, Holland_L)

solve_tulips_holland = sp.solve(equ_h, T)
# They are the same, so we can use the same variable to store the result
# So if beer is 33.33, then tulips will be 33.33 as well
solve_beer_holland = solve_tulips_holland

print(f'Belgium: {solve_beer_belgium}, Holland: {solve_beer_holland}', sep='\n')
```

In autarky Belgium produces 33.33 units of beer and 33.33 units of tulips, while Holland produces 66.67 units of beer and 66.67 units of tulips.

```python
#chatgpt garbage and i still dont know how to calculate the equilibrium price of beer in world trade, 
# I think it is 1 but I am not sure. I will try to calculate it later correct if I am wrong

# Define the symbol for relative price of beer to tulips
P_B = sp.Symbol('P_B', positive=True)  # Relative price of beer to tulips

# Constants
L_belgium = 100   # Labor in Belgium
L_holland = 200   # Labor in Holland
MPL_B = 1         # MPL for Beer in Belgium
MPL_T = sp.Rational(1, 2)  # MPL for Tulips in Belgium
MPL_B_star = sp.Rational(1, 2)  # MPL for Beer in Holland
MPL_T_star = 1       # MPL for Tulips in Holland

# Opportunity costs
OC_beer_belgium = MPL_T / MPL_B    # Opportunity cost of beer in Belgium (0.5)
OC_beer_holland = MPL_T_star / MPL_B_star  # Opportunity cost of beer in Holland (2)

# World Relative Supply function depending on the relative price
def world_relative_supply(price):
    if price < OC_beer_belgium:
        return 0  # No beer produced if price is less than Belgium's OC
    elif OC_beer_belgium <= price < OC_beer_holland:
        return (L_belgium * MPL_B) / (L_holland * MPL_T_star)  # Belgium produces beer, Holland produces tulips
    else:
        return float('inf')  # Both countries produce beer

# Relative demand is constant at 1 (1:1 ratio for Leontief preferences)
RD = 1

# Define price candidates between the opportunity costs
price_candidates = [0.5, 1, 1.5, 2, 3]  # Adjusted price candidates

equilibrium_price = None

for price in price_candidates:
    RS_value = world_relative_supply(price)
    # Check if RS matches RD
    if RS_value == RD:
        equilibrium_price = price
        break

# Display the equilibrium relative price of beer
if equilibrium_price is not None:
    print(f"Trade Equilibrium Relative Price of Beer: {equilibrium_price}")
else:
    print("No equilibrium price found within the given range.")
```
